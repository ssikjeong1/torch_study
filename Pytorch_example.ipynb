```
## API setting
import torch
from torch.autograd import Variable
```

```
## Define the input data
x_data = Variable(torch.Tensor([[1.], [2.], [3.]]))
y_data = Variable(torch.Tensor([[3.], [6.], [9.]]))
```

```
## Define the model
# In this case, we typically make a model using "class"
class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear = torch.nn.Linear(1, 1)
  
  def forward(self, x):
    y_pred = self.linear(x)
    return y_pred
```

```
model = Model()
```

```
## Loss function and optimizer
Loss = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```
## Training loop
for epoch in range(1000):
  # Forward pass
  y_pred = model(x_data)

  # Compute and describe loss
  loss = Loss(y_pred, y_data)
  print(epoch, loss.data)

  # Zero gradients, backward pass, update the weights
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```
## Define test input data
Test_x = Variable(torch.Tensor([[5.]]))
```

```
## Epoch 200, result:
print("Predict:", model.forward(Test_x).data)
```

```
## Epoch 500, result:
print("Predict:", model.forward(Test_x).data)
```
